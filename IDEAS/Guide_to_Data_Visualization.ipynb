### Beginners Guide To Data Visualization

```python
# Importing relevant libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Set the visuals style correctly

plt.style.use("default")
sns.set_theme()
```

```python
# Load eexample datasets from the seaborn library

tips_df = sns.load_dataset("tips")
flights_df = sns.load_dataset("flights")
iris_df = sns.load_dataset("iris")
```

### 1. COMPARISON PLOT: How do things compare to each other ?
#### 1A. Bar Chart or Plot

```python
# Bar Chart Example for x = 'day' and y = 'total_bill' from the "tips" dataframe. 
plt.figure(figsize=(10, 5))
sns.barplot(
    data=tips_df,
    x = 'day',
    y = 'total_bill',
    estimator=np.mean,
)

# Customizing our Bar Graph
plt.title("Which Days Have Higher Restaurant Bills ?")
plt.xlabel("Day of Week")
plt.ylabel("Average Bill ($)")

# Display Bar chart
plt.show()
```

#### 1B. Line Graph or Chart

```python
# Line Chart Example for x = 'Month' and y = 'No of passenger' from the "flight" dataframe.
plt.figure(figsize=(10, 5))

# fitering out from the entire Dataset, the data we will be using for our plot
monthly_flights = flights_df.pivot(index='month', columns='year', values= 'passengers')

# Getting our data from the variable "monthly_flight"
sns.lineplot(data=monthly_flights)

# customizing our plot
plt.title("How Does Air Travel change Throughout the Year ?")
plt.xlabel("Number of Passengers")
plt.show()
```

### 2. DISTRIBUTION PLOTS: "How are value spreed out ?"

#### 2A. Histogram

```python
# Histogram Example for x = 'Bill Amount ($)' and y = 'How many Times This Amount Occured' from the "tips" dataframe.
plt.figure(figsize=(10, 5))
sns.histplot(
    data=tips_df,
    x='total_bill',
    kde=True, # Margin line around the histogram
    bins=30 # size of the width standing bar
)


# Customizing plot
plt.title("How Common Are Different Bill Amounts ?")
plt.xlabel("Bill Amount ($)")
plt.ylabel("How Many Times This Amount Occured")

# Displaying plot
plt.show()
```

#### 2B. Box Plot

```python
# Box Plot Example for x = 'Day of the Week' and y = 'Bill Amount ($)' from the "tips" dataframe.
plt.figure(figsize=(10, 5))

sns.boxplot(
    data=tips_df, # Specifying where our data set will be gotten from.
    x='day', # selecting the data for the x-axis from the entire dataframe "tips_df"
    y='total_bill' # selecting the data for the y-axis from the entire dataframe "tips_df"
)

# Customizing plot
plt.title("Rnage of Resturant Bills by Day")
plt.xlabel("Day of week")
plt.ylabel("Bill of Amount ($)")

# Displaying plot
plt.show()
```

### 3. RELATIONSHIP PLOTS: "How do things affect each other ?"
#### 3A. Scatter Plot

```python
# scatter Plot Example for x = 'Total Bill ($)' and y = 'Tips Amount ($)' from the "tips" dataframe.

plt.figure(figsize=(10, 5))
sns.scatterplot(
    data=tips_df,
    x='total_bill',
    y='tip',
    hue='time',   # Decoration
    style='time' # Differentiate between respective dots.
)

plt.title("Do Bigger Bill Get Bigger Tips?")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip Amount ($)")

plt.show()
```

#### 3A. Correlation Heatmap

Corretion heatmap will be plotted on only numeric columns, the selected numeric columns will be coming from only the iris dataset or dataframe.

```python
# Selecting only numerical column from iris dataset
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# plotting correlation from numerical columns
correlation = iris_df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm', # color
    center=0,
    fmt='.2f' # Format correlation values to 2 decimal places
)

plt.title("How Strong Are Relationship Between Iris numerical Features?")
plt.show
```

### 4. COMPOSITION PLOTS: "How do parts make whole ?"
#### 4A. Pie Chart

```python
# Size of pie chart 
plt.figure(figsize=(8, 8))

# Obtaining data to be used for the plot and specifying how it will be used in the pie chart.

tips_df['time'].value_counts().plot(   # Count the value of lunch and dinner in the "time" and create a plot for your discovery.  
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("What Percentage of Customers Come for Lunch vs Dinner ?")
plt.ylabel('')
plt.show()
```

#### 4B. Stacked Bar

```python
# Obtaining data to be used for the plot and specifying how it will be used in the Stacked Bar.
time_day = pd.crosstab(tips_df['day'], tips_df['time'])
plt.figure(figsize=(10, 5))
time_day.plot(kind= 'bar', stacked=True)

# Customizing stacked Bar Chart
plt.title("Lunch vs Dinner Customer by Day")
plt.xlabel("Day of Week")
plt.ylabel("Number of Customers")
plt.legend(title="Meal Time")

# Displaying stacked Bar chart
plt.show()
```
