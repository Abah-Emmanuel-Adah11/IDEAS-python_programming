#### 3. SUMMARY STATISTICS

```python
# To calculate the summary statistics we may need to loop throught the numerical columns before proceeding to calculate the summary statistics.\

numerical_columns = ['bedrooms','bathrooms','toilets','parking_space','price'] # You can adjust as needed for your calculation.

for column in numerical_columns:
    print (f"\nSummary statistics for {column}:") # The "f" in the print statement ensures that string "summary....." and the content of the loop
                                                 # "column"can be printed from thesame print function without any error. Note that the item to be
                                                # printed must be covered within a curly brackek "{ }"
    
    print (f"Mean: {nigeria_houses[column].mean()}") # The loop "column" captures the title of the respective columns to be printed from the nigeria
                                                    # _houses data set. Hence, attaching them the .mean() function prints their respective means indiv-
                                                    # ually from the data set on the bases of how they are arranged in the loop.
    
    print (f"median: {nigeria_houses[column].median()}") # same thing applies to the median.
    
    print (f"Mode: {nigeria_houses[column].mode()[0]}") # same thing applies to the mode but the mode takes the first case in case of multiple columns.
                                                        # Hence, you can specify the specific column you want the calculation done. 

# Calcuting the range
    data_range = nigeria_houses[column].max() - nigeria_houses[column].min()
    print(f"Range: {data_range}")

# calculate variance and standard deviation
    print(f"Variance: {nigeria_houses[column].var()}")
    print(f"Standard Deviation: {nigeria_houses[column].std()}")
```

#### 4. DATA VISUALIZATION

#### 4.1 Univariate Analysis

##### Histograms

##### Countplots

```python
# To enable us visualize our Data set properly, so that we can differentiate between our numerical and categorical variables
# by printing the first five rows of the data set using the .head() function.

nigeria_houses.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Using Histogram to visualize the numerical columns

numerical_columns = ["bedrooms","bathrooms","toilets","parking_space","price"]
for column in numerical_columns:
    sns.histplot(data=nigeria_houses[column], kde=True, bins=20)
    plt.title(f"Distribution of {column}")   #This focuses on the title of the Distribution or Histogram.
    plt.xlabel(column)       # This consetrates on the data to be represented on the x-axis of the Histogram.
    plt.ylabel("frequency")  # This caputures the specification of the y-axis of the Histogram.
    plt.show()  # This is equivalent to the print() function, it direct's the interpreter to display the graph.

    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Using countplot for visualization of the categorical columns

categorical_columns = ["state","town"]
for column in categorical_columns:
    sns.countplot(y=nigeria_houses[column], order=nigeria_houses[column].value_counts().index)
    plt.title(f"frequency of {column}")
    plt.xlabel("Count")
    plt.ylabel(column)
    plt.show()
```

```python
import numpy as np

# log transformation of the price column to compress its extreme values
nigeria_houses["log_price"] = np.log1p(nigeria_houses["price"]) # converting price to its log form using the numpy library.

# Plotting the histogram of the log-transformed prices
plt.figure(figsize=(8,5))
plt.hist(nigeria_houses["log_price"], bins=50, edgecolor="black")
plt.title("Histogram of log-transformed Prices")
plt.xlabel("log(price)")
plt.ylabel("Frequency")
plt.show()
```

#### 4.2 Bivariate Analysis (Exploring relationship between variables)

#### Scatter Plots

```python
# Scatter plots for numerical variables versus price
for column in ["bedrooms","bathrooms","toilets","parking_space"]: # A for loop that iterate directly through all the numerical variables. 
                                                                # thus saving the result of this iteration in the "column" container.
    
    sns.scatterplot(x=nigeria_houses[column], y=nigeria_houses["price"]) # This gives the description of how the scatter plot should look
                                                                          # i.e the "column" cantainer should be on the y-axis & price on the x-axis.
                                                                        # In the plot.
    plt.title(f"price versus {column}") # The heading or title
    plt.xlabel(column) # the x-axis label
    plt.ylabel("price") # the y-axis label
    plt.show() # The command to display the plot.
    
```

#### Box Plots

```python
# Box plots for categorical variables versus price
for column in ["state", "town"]:
    sns.boxplot(x=nigeria_houses[column], y=nigeria_houses["price"])
    plt.title(f"Price versus {column}")
    plt.xlabel(column)
    plt.ylabel("price")
    plt.xticks(rotation=60) # This affect how the various items on the x-axis will be displayed. for instance, "rotation=45" will make the write-ups
                            # on the x-axis a kind of slanthy while "rotation=90" will make the write-ups display straight up. 
    plt.show()
```
