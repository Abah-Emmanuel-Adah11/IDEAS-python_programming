### INTRODUCTION TO Matplotlib
**Matplotlib** is a versatile library in python for creating visualization. such as:
1. Line plots
2. Bar charts

   And more, It is widely used for data analysis and scientific plotting.

**%matplotlib inline** magic command in jupyter Notebook ensures that plots are displayed directly on the notebook. Thus, making it easy to visualize data while coding.

On the other hand the **import matplotlib.pyplot as plt** command imports the ***pyplot module*** which provides a MATLAB-like interface for creating and customizing visualization in python. It simplifies tasks like ploting graphs and formating axes. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

**Plotting with Matplotlib**

Using the ***plt.plot()*** function you can create a simple line plot to visualize relationships between two sets of data e.g

```python
x = [1,2,3,4] # Independent variable values
y = [11, 22, 33, 44] # Dependent variable values

# The Command to create a line plot of y against x
plt.plot(x, y);
```

**Customizing Plots and Creating Subplots with Matplotlib**

You can create a figure and axes with specific dimensions using ***plt.subplots***. This allows you to control the size of the plot and manage multiple subplots, which are useful for comparing different visualizations side by side.

**What is a Subplot?**

A ***subplot*** refers to multiple smaller plots that are part of a larger figure. Using subplots you can display several charts in a grid layout within a single figure. This is important for comparing different datasets or visualizations in the same window.


```python
# Creating a figure with specific size i.e (figsize=(10,10) or figsize=(5,5) etc.) and plotting on it.

fig, ax = plt.subplots(figsize=(5,5)) # The first 5 rep the length of x-axis while the second 5 rep the length of the y-axis

# Plotting the x and y data on the axes
ax.plot(x,y);
```

 **Displaying multiple subplots in one line**

```python
# Code for displaying two subplots in one line

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6)); # Specifying how both subplots should look.
ax1.plot(x, y); # this line of code, plots the x against the y axis.
ax2.plot(y, x); # this line of code plots the y against the x axis.
```

```python
# How to Customize plot when and where necessary.
ax.set(title="simple plot",
       xlabel="x-aixs",
       ylabel="y-axis")

# saving the output of the plot above as an image file to the home page of our jupyter notebook as "first_visualization.png".
fig.savefig("first_visualition.png")
```

### Making figures with numpy arrays
**numpy** is a powerful python library for numerical computting.

**import** *numpy* ***as*** *np* is the magic command for importing the numpy library into your work.

The function **np.linspace(start, stop, num)** helps us generate an array of evenly spaced values over a specified range. It is commonly used to create a range of values to be used in plots or calculations.

For instance of we say **np.linespace(0, 10, 100)** we are simply instructing our numpy llibrary to create a series of **100** sequence **Begining** at zero **0** and **Ending** at ten **10**   

```python
# for instance
x = np.linspace(0, 10, 100)
x # printing out our result
```

```python
# Another example (personal)
y = np.linspace(2, 10, 5) # i.e create a series of 5 sequence begining at 2 and ending at 10
y # printing out our result
```

```python
# plotting a "line plot" of "x" against "exponentia x"

fig, ax = plt.subplots()
ax.plot(x, np.exp(x));
```

**Using python Dictionary to Generate Plots**

```python
# Dictionary for the prices of different types of nuts
nut_prices = {
    "Groundnut": 3000,
    "cashew Nut": 10000,
    "Almounds": 8000,
    "Pistachio": 12000,
    "Walnut": 15000
}

# Displaying our python dictionary on a bar chat
fig, ax = plt.subplots()
ax.bar(nut_prices.keys(), nut_prices.values());

# customizing our title and ylabel
ax.set(title="Nigeria nut prices 2024",
    ylabel="Prices"
);
```

#### **Subplots: Object Oriented Methods**

##### Using object oriented method to plot 2 rows and 2 columns of subplots in a grid format.

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, # specifying the numbers of rows
                                            ncols=2,   # specifying the numbers of columns
                                            figsize=(10,5)) # specifying the size of the figures 
```

**Customizing Subplots Plotted using object oriented method**

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))  # Specifying how the subplots should look.


x = np.linspace(0, 10, 100)        # Obtaining our figure using numpy array.
ax1.plot(x, x / 2)              # Specifying that our x be plotted against x / 2.
ax1.set_title("Line plot");    # customizing the title of our plot which will be a line plot by default.


ax2.scatter(np.random.random(10), np.random.random(10), color='r') # Using Numpy to generate 10 set of random numbers and also customizing our color to red
ax2.set_title("Scatter Plot");


prices = {"Groundnut": 3000, "Cashew nut": 10000, "Almond": 7000, "walnut": 5000}  # plotting from a python dictionary, just as in previous examples
ax3.bar(prices.keys(), prices.values(), color='g')
ax3.set_title("Bar Chart")
ax3.set_ylabel("Price (Naria)");

ax4.hist(np.random.random(1000), bins=100, color='b') # the "bins=100" is used to specify the sizes of the vertical bars. It can be incresed or decreased depending on your preference
ax4.set_title("Histogram")
plt.show;
```

#### **Understanding Bins in Data Visualization**

**Bins** are intervals used to group data point in visualization like Histogram.
* **Purpose:** Simplifies distributions for easy visualization.
* **Key Points:**
 
     ***Fewer bins:*** Less detailed intervals.
  
     ***More bins*** Narrow more detailed but can add noise.

* **Example:** python plt.hist(data, bins=20)

#### **Plotting from Dataframe**

```python
# Importing data
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

**NB:** If we go ahead to plot directly like so **car_sales.plot(x="Make", y="Price")** it will throw an error because the computer will
be interpreting the the numeric data as a string. Hence we need to convert our said string (Price) to an integer or float using the **astype()** function. Depending on the type of data. 

```python
car_sales["Price"] = car_sales["Price"].str.replace("$", "").str.replace(",","").astype(float) 
# Aside the .astype() we also included the .str.replace() function to replace the special characters ($ and ,) in our price data.
# After this we can now proceed to plot our graph using the same codes we were to use ealier.

car_sales.plot(x="Make", y="Price");
```

**NB:** Using the object oriented method, using the code **car_sales.scatter(x="Make", y="Price")** to generate a scatter plot will throw an error, since the 'DataFrame' object **does not recognize the scatter attribute**. To achive the above we might need to modify the code as follows: **car_sales.plot(x="Make", y="Price", kind="scatter");**

```python
car_sales.plot(x="Make", y="Price", kind="scatter");
```

```python
# Another example (personal). using same data set to plot a Bar chart.
car_sales.plot(x="Make", y="Price", kind="bar");
```

### **Some data analysis on our nigeria houses dataset with Visualizations**

```python
# importing the data to jupyter notebook
nigeria_houses = pd.read_csv("nigeria_houses_data.csv")
nigeria_houses.head()
```

```python
# Getting the data type of our "Price" column.
nigeria_houses["price"].dtypes
```

```python
# Calculating the mean of the "price" column.
nigeria_houses["price"].mean()
```

```python
# sorting out the prices that are greater than over 1 billion.
over_1bn = nigeria_houses[nigeria_houses["price"] > 1000000000.0]
over_1bn.head()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10)) # setting the templaates for the plots

# creating a subplot and taking the log to base 10 of the "price" column that are over 1 billion to address the issue of outliers in our data set.
# The prices that are over 1 billion will be plotted against state.


# scatter plot 1 (prices over 1bn against state)
scatter_plot = ax0.scatter(x= np.log10(over_1bn["price"]), y=over_1bn["state"])
# customizing the plot
ax0.set_title("Log10 of prices by state")
ax0.set_xlabel("Log10(price in Naira)")
ax0.set_ylabel("State")
ax0.legend(["Log-transformed Prices"], loc="upper right");   # this gives a kind of subtitle and also shows the keys used. As can be seen from the diagram and  key


# Scatter plot 2 (prices over 1bn against town)
scatter_plot2 = ax1.scatter(x = np.log10(over_1bn["price"]), y= over_1bn ["town"])

# customizing the plot
ax1.set_title("Log10 of prices by Town")
ax1.set_xlabel("Log10(Price in Naira)")
ax1.set_ylabel("Town")
ax1.legend(["Log-transformed Prices"], loc="upper right");
```
